# Tic Tac Toe

```python
the_board = {
    1: ' ', 2: ' ', 3: ' ',
    4: ' ', 5: ' ', 6: ' ',
    7: ' ', 8: ' ', 9: ' '
}

def print_board(board):
    print(f"{board[1]} | {board[2]} | {board[3]}")
    print("- + - + -")      
    print(f"{board[4]} | {board[5]} | {board[6]}")
    print("- + - + -")      
    print(f"{board[7]} | {board[8]} | {board[9]}")   
  
print_board(the_board)

def check_end_game(board):
    if board[1] == board[2] == board[3] == 'X' or \
    board[4] == board[5] == board[6] == 'X' or \
    board[7] == board[8] == board[9] == 'X':
        print("X won the game.")
        return True
        
    elif board[1] == board[4] == board[7] == 'X' or \
    board[2] == board[5] == board[8] == 'X' or \
    board[3] == board[6] == board[9] == 'X':
        print("X won the game.")
        return True
        
    elif board[1] == board[5] == board[9] == 'X' or \
         board[3] == board[5] == board[7] == 'X':
        print("X won the game.")
        return True
        
    elif board[1] == board[2] == board[3] == 'O' or \
    board[4] == board[5] == board[6] == 'O' or \
    board[7] == board[8] == board[9] == 'O':
        print("O won the game.")
        return True
        
    elif board[1] == board[4] == board[7] == 'O' or \
    board[2] == board[5] == board[8] == 'O' or \
    board[3] == board[6] == board[9] == 'O':
        print("O won the game.")
        return True
        
    elif board[1] == board[5] == board[9] == 'O' or \
         board[3] == board[5] == board[7] == 'O':
        print("O won the game.")
        return True
        
    else: 
        return False
    
def check_all_filled(board):
    if board[1] != ' ' and board[2] != ' ' and board[3] != ' ' \
    and board[4] != ' ' and board[5] != ' ' and board[6] != ' ' \
    and board[7] != ' ' and board[8] != ' ' and board[9] != ' ':
        return True

turn = 'X'
while True:
    print(f"Turn for {turn}. Move on which space(1-9)?")
    move = int(input())
    if the_board[move] == ' ':
        the_board[move] = turn
    else:
        print("Space is occupied. Choose another: ")
        continue
    if turn == 'X':
        turn = 'O'
    else: 
        turn = 'X'
    print_board(the_board)
    win = check_end_game(the_board)
    if win == True:
        break
    else:
        if check_all_filled(the_board):
            print("No one won the game.")
            break
            
        else:
            continue
```
