## Roman Alphabets Into Number

```python
### Convert Roman Number alphabets into Number
import sys

Roman_list = list(input("Enter Roman string:").upper())
# Roman_list = list(Roman)
print(Roman_list)
Number_list = []
for i in range(len(Roman_list)):
    case = Roman_list[i]
    if case == 'I':
        e = 1
        Number_list.append(e)
    elif case == 'V':
        e = 5
        Number_list.append(e)
    elif case == 'X':
        e = 10
        Number_list.append(e)
    elif case == 'L':
        e = 50
        Number_list.append(e)
    elif case == 'C':
        e = 100
        Number_list.append(e)
    elif case == 'D':
        e = 500
        Number_list.append(e)
    elif case == 'M':
        e = 1000
        Number_list.append(e)
    else:
        print("Enter Valid Roman Number")
        sys.exit()
# print(Number_list)
if len(Number_list) == 1 and (Number_list[0] == 1 or Number_list[0] == 5 or Number_list[0] == 10 or Number_list[0] == 50 or Number_list[0] == 100 or Number_list[0] == 500 or Number_list[0] == 1000):
    print(f"The Numeric Value of {Roman_list} is {Number_list[0]}")
    sys.exit()
else:
    before = Number_list[0]
    current = Number_list[1]

    for i in range(2, len(Number_list) ):
        if before < current and current <= Number_list[i]:
            print("Roman Numerals contain more than one higher Roman character after the lower one.")
            sys.exit()
        else:
            before = current
            current = Number_list[i]
    count = 0
    for i in range(1, len(Number_list) ):
        if Number_list[i-1] == Number_list[i]:
            count = count + 1
            if count >= 3:
                print("Roman Number contains more than three consecutive Roman letter in the combination.")
                sys.exit()
        else:
            count = 0

    current = Number_list[0]
    result = current
    for i in range(1, len(Number_list)):
        if current >= Number_list[i]:
            result = result + Number_list[i]
            current = Number_list[i]
            print(result)
        else:
            result = result + Number_list[i] - 2 * current
            current = Number_list[i]
            print(result)

    print(f"The Numeric value of {Roman_list} is {result}")
```
